Problem: 402, Strength of Materials (Fourth Edition) by Andrew Pytel and Ferdinand L. Singer. Harper International Edition. 

```python
from indeterminatebeam import Beam, Support, PointLoadV, TrapezoidalLoadV

beam = Beam(10)
```

```python
#changing units 

beam.update_units(key='length', unit='ft')
beam.update_units('force', 'lbf')
beam.update_units('distributed', 'lbf/ft')
beam.update_units('moment', 'lbf.ft')
```

```python
support = Support(10, (1,1,1))

beam.add_supports(support)
```

```python
load_1 = TrapezoidalLoadV((0,-60), (0,6))   #TrapezoidalLoadV((Initial load, Final load), (Initial distance, Final distance))
load_2 = PointLoadV(-200, 8)

beam.add_loads(load_1, load_2)
```

```python
fig = beam.plot_beam_diagram()
fig.show()
```

```python
beam.analyse()
```

```python
fig = beam.plot_beam_external()
fig.show()
```
