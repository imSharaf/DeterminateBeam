```python
from indeterminatebeam import Support, Beam, PointTorque, PointLoadV, PointLoad
%matplotlib inline

beam = Beam(18)
```

```python
beam.update_units('force','lbf')
beam.update_units('moment', 'lbf.ft')
beam.update_units('length', 'ft')
```

```python
a = Support(0,(1,1,0))
b = Support(18,(0,1,0))

beam.add_supports(a,b)
```

```python
load_1 = PointLoad(1800,6,53.13)
load_2 = PointTorque(-14400,12)

beam.add_loads(load_1, load_2)
```

```python
fig = beam.plot_beam_diagram()
fig.show()
```

```python
beam.analyse()
```

```python
fig = beam.plot_beam_external()
fig.show()
```

```python
fig = beam.plot_beam_internal()
fig.show()
```
