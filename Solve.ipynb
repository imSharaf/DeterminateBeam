```python
from indeterminatebeam import Support, Beam, PointTorque, PointLoadV

beam_1 = Beam(9)
beam_2 = Beam(3)
```

```python
beam_1.update_units('force','kN')
beam_1.update_units('moment', 'kN.m')

beam_2.update_units('force','kN')
beam_2.update_units('moment', 'kN.m')
```

```python
a = Support(0,(0,1,0))
b = Support(3,(1,1,0))

beam_2.add_supports(a,b)
```

```python
load_1 = PointLoadV(-4,1)

beam_2.add_loads(load_1)
```

```python
fig = beam_2.plot_beam_diagram()

fig.show()
```

```python
beam_2.analyse()
```

```python
fig = beam_2.plot_beam_external()

fig.show()
```

```python
a = Support(0,(1,1,0))
b = Support(6,(0,1,0))

beam_1.add_supports(a,b)
```

```python
load_1 = PointLoadV(-6,3)
load_2 = PointTorque(9,3)
load_3 = PointLoadV(-beam_2.get_reaction(0)[1],9)

beam_1.add_loads(load_1,load_2,load_3)
```

```python
fig = beam_1.plot_beam_diagram()
fig.show()
```

```python
beam_1.analyse()
```

```python
fig = beam_1.plot_beam_external()
fig.show()
```

```python
fig = beam_1.plot_beam_internal()
fig.show()
```
