```python
#Getting dependencies
from indeterminatebeam import Beam, Support, PointLoadV

beam = Beam(6)  #Creating an object of our desired beam with its length in m
```

```python
#adding supports 

a = Support(1, (1,1,0))   # Support (placement of support from left, (1=Horizonatal Reaction, 1=Vertical Reaction, 1=Moment))
b = Support(6, (0,1,0))   # Support(placement of support,(support type), kx,ky)

beam.add_supports(a,b)   
```

```python
#defining loads

load_1 = PointLoadV(-30000, 0)   # PointLoadV --> Vertical Load((load direction with sign)Load value in N, load position)
load_2 = PointLoadV(-50000, 4) 

beam.add_loads(load_1, load_2)   
```

```python
'''

Analyse the beam: computes all support reactions, shear forces, and bending moments
based on the defined supports and applied loads. Must be called before plotting SFD/BMD
for querying internal forces/moments.

'''

beam.analyse()
```

```python
#show reaction forces

fig_1 = beam.plot_beam_external()
fig_1.show()
```

```python
#show Normal Force, SFD, BMD and Deflection

fig_2 = beam.plot_beam_internal()
fig_2.show()
```
