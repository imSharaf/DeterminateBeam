Import and Access

```python
#import library
from anastruct import SystemElements

#object creation to access methods

ss = SystemElements()
```

```python
#adding elements

ss.add_element(location=[[0,0],[5,6]])  #beam element location--> location=[[start],[end]]
ss.add_element(location=[[5,6],[10,6]])

#adding supports 

ss.add_support_hinged(node_id=1)  #node id refers to the each node of frames. from left to right this increases from 1.
ss.add_support_roll(node_id=3, direction=2)   #direction refers to which direction the reaction is.

#adding loads

ss.q_load(element_id=2, q=-15)  #element id refers to the define number of the element. q refers the distributed load magnitude and sign for direction. 
```

```python
#initiate solve method to work on

ss.solve()
```
