```python
from indeterminatebeam import Beam, Support, PointLoadV, UDLV, PointTorque

beam = Beam(8)
```

```python
#changing units 

beam.update_units(key='length', unit='ft')
beam.update_units('force', 'lbf')
beam.update_units('distributed', 'lbf/ft')
beam.update_units('moment', 'lbf.ft')
```

```python
a = Support(2, (1,1,0))  #In question it is a roller support but in reality there is a x restraint too. However it will be zero. 
b = Support(8, (0,1,0))

beam.add_supports(a,b)
```

```python
load_1 = UDLV(-100, (0,6))     #UDLV (Load, (Start, End))
load_2 = PointTorque(-1200, 7)   #Moment--> PointTorque(+AntiClockwise Moment, Distance)

beam.add_loads(load_1, load_2)
```

```python
fig = beam.plot_beam_diagram()
fig.show()
```

```python
beam.analyse()
```

```python
fig = beam.plot_beam_external()
fig.show()
```

```python
fig = beam.plot_beam_internal()
fig.show()
```
